```python
import cv2
from google.colab.patches import cv2_imshow
def Face_identification (image):

  face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
  img = cv2.imread(image)
  gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
  faces = face_cascade.detectMultiScale(gray, 1.2,7)
  for (x, y, w, h) in faces:
    cv2.rectangle(img, (x,y), (x+w, y+h),(10, 20, 255), 2)
  cv2_imshow(img)
  cv2.waitKey()

```

```python
img = ['pexels-photo-3184398.webp',
       'Woman-with-soft-round-face-wearing-square-eyeglasses-frame.jpg',
       'Beautiful-smile.jpg',
       '360_F_157125888_DLSe6hsftUm2RpVENNqtC5il1gX2OIJR.jpg']
for i in img:
  Face_identification(i)
```
